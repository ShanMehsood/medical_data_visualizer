```python
# imports required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# now read the dataset 
data = pd.read_csv('medical_examination.csv')
```

```python
data
```

```python
#lets copy the oroiginal data 
df = data.copy()
```

```python
df
```

```python
# now add the column name overweight
df['overweight']= np.where(df['weight']/((df['height']*0.01)**2) >25, 1,0)
```

## This part calculates the BMI for each row in the DataFrame. The BMI formula is:

## BMI= weightÂ (kg)/(heightÂ (m))2

```python
df
```

```python
# normalize the data by 0  always good and 1 always bad ,.. if the value of cholestrol and gluc is 1 make it zero (0) ,if the 
# values is more then 1 make it 1

df['cholesterol']= np.where(df['cholesterol']==1 ,0,1)
```

```python
df['gluc']= np.where(df['gluc']==1 ,0,1)
```

```python
df
```

```python
# create DataFrame for cat plot using pd.melt
df_cat = sorted(['cholesterol','gluc','smoke' ,'alco','active','overweight'])
# here we have use sorted function to make  the list i ascending order
```

```python
df_cat
```

```python
df_cat = pd.melt(df,id_vars='cardio',value_vars = df_cat)
#This code uses the pd.melt function from the pandas library to transform the DataFrame df from a wide format to a long format.
# "melts" the DataFrame, turning columns into rows.
```

```python
df_cat
```

```python
#Draw catplot with sns.catplot()
sns.catplot(x ='variable',col ='cardio',hue = 'value',kind = 'count', data= df_cat).set_axis_labels('variable','total')
```

```python
#clean the data
df_heat = df[
    (df['ap_lo'] <=df['ap_hi']) &
    (df['height'] >=df['height'].quantile(0.025)) &
    (df['height'] <=df['height'].quantile(0.975)) &
    (df['weight'] >=df['weight'].quantile(0.025)) &
    (df['weight'] <=df['weight'].quantile(0.975))]
```

```python
corr = df_heat.corr(method = 'pearson')
```

```python
mask = np.triu(corr)
```

```python
#setup the matplotlib figure
fig ,ax = plt.subplots(figsize=(12,12)) 
```

```python
sns.heatmap (corr, linewidths = 1,annot = True,square = True , mask=mask ,fmt = ".1f",
center = 0.08,cbar_kws ={"shrink":0.5})
```

```python
fig.savefig('heatmap.png')
```
